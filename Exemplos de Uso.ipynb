```python
from highbond_sdk import HighBondClient
```

```python
# Inicializar cliente
client = HighBondClient(
    token="seu token_aqui",  # str - Token de API
    org_id=123456,        # int - ID da organização
    region="us"          # "us", "eu", "au", "ca" ou "sa"
)
```

# Projetos

```python
# Listar todos os projetos (paginação automática - generator)
for projeto in client.projects.list_all():
    print(f"- {projeto['id']}: {projeto['attributes']['name']}")
```

```python
# Buscar projeto específico
projeto = client.projects.get(546355, return_pandas=True)
projeto
```

```python
# Atualizar projeto
client.projects.update(541532, name="projeto draft", status="active")
```

```python
# Listar tipos de projeto
tipos_projetos = client.projects.list_project_types()
tipos_projetos
```

# Riscos 

```python
# Listar riscos de um projeto
client.risks.list_by_project(project_id=546355, return_pandas=True)
```

```python
# Listar riscos de um objetivo
client.risks.list_by_objective(objective_id=1990741, return_pandas=True)
```

```python
# Listar todos riscos de uma organização
client.risks.list_all(return_pandas=True)
```

```python
# Buscar risco
client.risks.get(risk_id=8496485, return_pandas=True)
```

```python
riscos = client.risks.list_all(return_pandas=True)['id']

# Buscar múltiplos em paralelo
client.risks.get_many(riscos, return_pandas=True)
```

```python
# Criar risco (dentro de um objetivo)
client.risks.create(
    objective_id=1990761,
    description="Descrição detalhada do risco",
    title="Título do Risco",
    impact="Medium",
    likelihood="Low",
    owner="responsavel@empresa.com",
    risk_id = 2
)
```

```python
# Atualizar risco
client.risks.update(risk_id=8496485, impact="High", title="Novo Título para o Risco", custom_attributes=[
    {'term': 'Uma possivel resposta como atributo', 
     'value': ['Essa é minha nova resposta']},
    {'term': 'Termo de teste',
     'value': ['Opção 1']}
     ])
```

```python
risco = client.risks.list_by_objective(objective_id=1990761, return_pandas=True)['id'].iloc[-1]

# Deletar risco
client.risks.delete(risk_id=risco)
```
